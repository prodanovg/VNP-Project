```python
# ChatOpenAI.model_rebuild()
# ChatOpenAI.BaseCache = BaseCache    
with open("output/youtubeVideo.txt", "r", encoding="utf-8") as f:
    input_text = f.read()

with open("config.yml", "r") as f:
    config = yaml.safe_load(f)

llm_config = config['llm']

if llm_config['model_type'] == "openai":
    llm = ChatOpenAI(
        model=llm_config.get('model_name', 'gpt-3.5-turbo'),
        temperature=llm_config.get('temperature', 0.7),
        max_tokens=llm_config.get('max_tokens', 2000),
        api_key=llm_config['api_key']
    )
else:
    raise ValueError("Only 'openai' model_type is implemented in this snippet.")

embeddings = OpenAIEmbeddings(api_key=llm_config['api_key'])
print("LLM initialized:", llm)
```

```python
distiller = DocumentsDistiller(llm_model=llm)
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
chunks = splitter.split_text(input_text)
itext2kg = iText2KG(llm_model=llm, embeddings_model=embeddings)


class Triple(BaseModel):
    subject: str
    predicate: str
    object: str


class TriplesOutput(BaseModel):
    triples: List[Triple]


all_triples = []
for i, chunk in enumerate(chunks, 1):
    try:
        kg = itext2kg.build_graph(
            sections=[chunk],
            ent_threshold=0.6,
            rel_threshold=0.6,
            output_data_structure=TriplesOutput

        )
        triples = [(rel.startEntity.name, rel.name, rel.endEntity.name) for rel in kg.relationships]
        all_triples.extend(triples)
        print(f"Chunk {i}/{len(chunks)} → {len(triples)} triples")
    except Exception as e:
        print(f"Chunk {i} failed:", e)
    time.sleep(5)

print("Total triples:", len(all_triples))
```

```python
triples_objects = [Triple(subject=s, predicate=p, object=o) for s, p, o in all_triples]
all_triples_model = TriplesOutput(triples=triples_objects)

semantic_blocks = [
    f"{triple.subject} {triple.predicate} {triple.object}"
    for triple in all_triples_model.triples
]

sections = [semantic_blocks]
```

```python
kg = itext2kg.build_graph(
    sections=sections,
    ent_threshold=0.6,
    rel_threshold=0.6
)
```

```python
G = itext2kg.graph

plt.figure(figsize=(12, 8))
nx.draw_networkx(G, with_labels=True, node_size=1500, font_size=10)
plt.axis('off')
plt.show()
```
