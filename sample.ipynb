```python
    # ChatOpenAI.model_rebuild()
    # ChatOpenAI.BaseCache = BaseCache
    # ChatOpenAI.model_rebuild()
    
    with open("youtubeVideo.txt", "r", encoding="utf-8") as f:
        input_text = f.read()
    
    with open("config.yml", "r") as f:
        config = yaml.safe_load(f)
    
    llm_config = config['llm']
    
    if llm_config['model_type'] == "openai":
        llm = ChatOpenAI(
            model=llm_config.get('model_name', 'gpt-3.5-turbo'),
            temperature=llm_config.get('temperature', 0.7),
            max_tokens=llm_config.get('max_tokens', 2000),
            api_key=llm_config['api_key']
        )
        llm.model_rebuild()  
    else:
        raise ValueError("Only 'openai' model_type is implemented in this snippet.")
    
    
    embeddings = OpenAIEmbeddings(api_key=llm_config['api_key'])
    print("LLM initialized:", llm)
```

```python
class Triple(BaseModel):
    subject: str
    predicate: str
    object: str

class TriplesOutput(BaseModel):
    triples: List[Triple]

IE_query = '''
- Extract subject, predicate, object triples from the following text chunk.
- Return output strictly following the TriplesOutput Pydantic schema.
'''

input_text_chunks = [input_text] 

document_distiller = DocumentsDistiller(llm_model=llm)
distilled_triples = document_distiller.distill(
    documents=input_text_chunks,
    IE_query=IE_query,
    output_data_structure=TriplesOutput,
)


print("Distilled Sentences (Sample):\n")
for triple in distilled_triples.triples:
    print(triple.subject, triple.predicate, triple.object)

```

```python
itext2kg = iText2KG(
    llm_model=llm,
    embeddings_model=embeddings,
)
```

```python
semantic_blocks = [
    f"{triple.subject} {triple.predicate} {triple.object}"
    for triple in distilled_triples.triples
]

sections = [semantic_blocks]
```

```python
kg = itext2kg.build_graph(
    sections=sections,
    ent_threshold=0.6,  
    rel_threshold=0.6
)
```

```python
G = kg.graph  

plt.figure(figsize=(12, 8))
nx.draw_networkx(G, with_labels=True, node_size=1500, font_size=10)
plt.axis('off')
plt.show()
```
